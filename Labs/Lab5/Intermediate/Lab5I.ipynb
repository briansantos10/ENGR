```python
# Import libraries
import pandas as pd
import numpy as np
from scipy.integrate import trapezoid
import matplotlib.pyplot as plt

# --- Load CSV ---
df = pd.read_csv("motion_b_lab5.csv")

# Preview the data
print(df.head())

# --- Compute statistics ---
mean_distance = df['distance_cm'].mean()
median_distance = df['distance_cm'].median()
std_distance = df['distance_cm'].std()

print("Mean Distance:", mean_distance)
print("Median Distance:", median_distance)
print("Standard Deviation:", std_distance)

# --- Compute area under the curve (distance vs time) ---
area = trapezoid(df['distance_cm'], df['time_ms'])
print("Trapezoid Area under curve:", area)

# --- Plot distance vs time ---
plt.figure(figsize=(10,5))
plt.plot(df['time_ms'], df['distance_cm'], label='Distance (cm)', marker='o', linestyle='-')
plt.axhline(y=mean_distance, color='r', linestyle='--', label='Mean Distance')
plt.xlabel("Time (ms)")
plt.ylabel("Distance (cm)")
plt.title("Cutebot Distance Over Time")
plt.legend()
plt.grid(True)
plt.tight_layout()

# --- Save plot ---
plt.savefig("motion_b_plot.png", dpi=300)
plt.show()
```
