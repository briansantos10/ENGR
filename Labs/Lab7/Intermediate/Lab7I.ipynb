```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("data_b.csv")

time_s = df["time_ms"].values / 1000.0  # convert ms to s
position = df["distance_cm"].values

velocity = np.gradient(position, time_s)

acceleration = np.gradient(velocity, time_s)

df["velocity"] = velocity
df["acceleration"] = acceleration

fig, axs = plt.subplots(3, 1, figsize=(10, 8), sharex=True)

axs[0].plot(time_s, position, color="blue")
axs[0].set_ylabel("Distance (cm)")
axs[0].set_title("Distance vs Time")

axs[1].plot(time_s, velocity, color="green")
axs[1].set_ylabel("Velocity (cm/s)")
axs[1].set_title("Velocity vs Time")

axs[2].plot(time_s, acceleration, color="red")
axs[2].set_ylabel("Acceleration (cm/sÂ²)")
axs[2].set_xlabel("Time (s)")
axs[2].set_title("Acceleration vs Time")

plt.tight_layout()
plt.savefig("motion_b_analysis.png")
plt.show()
```
