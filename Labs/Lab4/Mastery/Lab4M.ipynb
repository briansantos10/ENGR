```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load data
df = pd.read_csv("lab4_a.csv")

# Quick sanity check
df.head()
```

```python
# Controller ran every 100 ms, so time increases by 0.1 seconds per row
df["time"] = df.index * 0.1
```

```python
plt.figure()
plt.plot(df["time"], df["distance"], label="Distance (cm)")
plt.plot(df["time"], df["error"], label="Error (cm)")

plt.xlabel("Time (s)")
plt.ylabel("Value")
plt.title("Distance and Error vs Time")
plt.legend()
plt.grid(True)

plt.savefig("plot_a.png")
plt.show()
```

```python
plt.figure()
plt.plot(df["time"], df["velocity"], label="Velocity Command")

plt.xlabel("Time (s)")
plt.ylabel("Velocity")
plt.title("Velocity Command vs Time")
plt.legend()
plt.grid(True)

plt.savefig("plot_a.png")
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Distance + Error
plt.subplot(2, 1, 1)
plt.plot(df["time"], df["distance"], label="Distance (cm)")
plt.plot(df["time"], df["error"], label="Error (cm)")
plt.ylabel("Value")
plt.title("Distance and Error vs Time")
plt.legend()
plt.grid(True)

# Velocity
plt.subplot(2, 1, 2)
plt.plot(df["time"], df["velocity"], label="Velocity Command")
plt.xlabel("Time (s)")
plt.ylabel("Velocity")
plt.title("Velocity Command vs Time")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.savefig("plot_a.png")
plt.show()
```
