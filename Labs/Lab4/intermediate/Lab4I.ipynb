```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("lab4_b.csv")

df.head()
```

```python
plt.figure(figsize=(8, 4))
plt.plot(df.index, df["distance_cm"], label="Distance (cm)")

plt.xlabel("Time (samples)")
plt.ylabel("Distance (cm)")
plt.title("Distance Over Time")
plt.legend()
plt.grid(True)

plt.savefig("plot_b_distance.png")
plt.show()
```

```python
plt.figure(figsize=(8, 4))

plt.plot(df.index, df["accel_x"], label="Acceleration X")
plt.plot(df.index, df["accel_y"], label="Acceleration Y")
plt.plot(df.index, df["accel_z"], label="Acceleration Z")

plt.xlabel("Time (samples)")
plt.ylabel("Acceleration")
plt.title("Acceleration Over Time")
plt.legend()
plt.grid(True)

plt.savefig("plot_b_acceleration.png")
plt.show()
```
