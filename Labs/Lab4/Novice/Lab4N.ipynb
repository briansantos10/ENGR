```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Lab4_c.csv")
data.head()
```

```python
plt.figure(figsize=(9, 5))
plt.plot(data["Time (seconds)"], data["light"], label="Light Level")
plt.plot(data["Time (seconds)"], data["temp"], label="Temperature (Â°C)")
plt.xlabel("Time (seconds)")
plt.ylabel("Sensor Values")
plt.title("Light and Temperature vs Time")
plt.legend()
plt.grid(True)

plt.savefig("plot_c.png")
plt.show()
```
