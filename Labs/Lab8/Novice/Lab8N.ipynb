```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("data_c.csv")

df["label"] = df["label"].str.strip().str.lower()

df.head()
```

```python
df["label"].value_counts()
```

```python
plt.figure(figsize=(10, 6))

for label in df["label"].unique():
    subset = df[df["label"] == label]
    plt.hist(subset["value"], bins=20, alpha=0.6, label=label)

plt.xlabel("Sound Level")
plt.ylabel("Count")
plt.title("Microphone Sound Level Distribution by Class")
plt.legend()
plt.grid(True)
plt.show()
```
